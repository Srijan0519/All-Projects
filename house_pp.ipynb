# House Price Prediction using Turicreate library

# Launch Turi Create

```python
import turicreate
```

# Load house sales data

```python
sales = turicreate.SFrame.read_csv
```

```python
sales
```

# Explore

```python
sales.show()
```

```python
turicreate.show(x=sales[1:5000]["price"],y=sales[1:5000]["bedrooms"]) 
```

## TRAIN AND TEST SPLIT

```python
train_data, test_data = sales.random_split(0.8,seed=0)
```

## FINDING THE BEST LINEAR REGRESSION MODEL 

```python
bedroom_model=turicreate.linear_regression.create(train_data, target="price",features=["bedrooms"])
```

## Evaluate the quality of our model

```python
print (test_data['price'].mean())
```

```python
print (bedroom_model.evaluate(test_data))
```

# Explore model a little further

```python
bedroom_model.coefficients
```

## USING MATPLOTLIB FOR FURTHER VISUALISATION AND PREDICTION

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

try:
    plt.plot(test_data["bedrooms"],test_data["price"],".",
         test_data["bedroom_model"],bedroom_model.predict(test_data),"-")
except:
    plt.plot(test_data["bedrooms"],test_data["price"],".",
         test_data["bedrooms"],bedroom_model.predict(test_data),"-")
```

# Explore other features of the data

```python
my_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
```

```python
sales[my_features].show()
```

```python
turicreate.show(sales['zipcode'],sales['price'])
```

# Build a model with these additional features

```python
my_features_model = turicreate.linear_regression.create(train_data,target='price',features=my_features)
```

# Compare simple model with more complex one

```python
print (my_features)
```

```python
print (bedroom_model.evaluate(test_data))
print (my_features_model.evaluate(test_data))
```
